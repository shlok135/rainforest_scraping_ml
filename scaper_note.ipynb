```python
base_url = "https://www.wunderground.com/history/monthly/br/v%C3%A1rzea-grande/SBCY/date/"

# we are only trying this for one year
years = list(range(1999,2020))
months = list(range(1,13))

list_of_dates = []
for year in years:
  for month in months:
    date = F"{year}-{month}"
    print(date)
    list_of_dates.append(date)

columns_for_df = ['max_temperature_max_month',
                  'max_temperature_avg_month',
                  'max_temperature_min_month',
                  'avg_temperature_max_month',
                  'avg_temperature_avg_month',
                  'avg_temperature_min_month',
                  'min_temperature_max_month',
                  'min_temperature_avg_month',
                  'min_temperature_min_month',
                  'dew_point_max_month',
                  'dew_point_avg_month',
                  'dew_point_min_month',
                  'precipitation_max_month',
                  'precipitation_avg_month',
                  'precipitation_min_month',
                  'precipitation_sum_month',
                  'snow_depth_max_month',
                  'snow_depth_avg_month',
                  'snow_depth_min_month',
                  'snow_depth_sum_month',
                  'wind_max_month',
                  'wind_avg_month',
                  'wind_min_month',
                  'gust_wind_max_month',
                  'gust_wind_avg_month',
                  'gust_wind_min_month',
                  'sea_level_pressure_max_month',
                  'sea_level_pressure_avg_month',
                  'sea_level_pressure_min_month',
                  'date',
                  'location',
                  ]
# Set up options for the Chrome webdriver
chrome_options = Options()
# chrome_options.add_argument("--headless")  # Run in headless mode (no GUI)
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model

# Set up the Chrome webdriver
driver = webdriver.Chrome(options=chrome_options)

building_df = pd.DataFrame(columns =columns_for_df)
row_counter = 0
list_of_dfs = []
for date in list_of_dates:
    temp_url = base_url + date
    # Navigate to the web page you want to scrape

    imported_flag = False
    failed_atempts = 0
    while imported_flag == False and failed_atempts < 5:
      try:
        driver.get(temp_url)

        # Get the page source and pass it to Beautiful Soup for parsing


        sleep(3)

        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")


        table_div = soup.find_all(class_="summary-table")


        # Do your scraping here...
        print(table_div)

        table_for_df = table_div[0].find("table")

        # if this does not work it will throw an error
        df = pd.read_html(str(table_for_df))[0]

        imported_flag = True
        print(F"Successful import for date {date} moving into next season ")
      except ImportError:
        failed_atempts = failed_atempts + 1
        print(F"Failed to import date {date} attempt {failed_atempts}")


    # print(df.head())

    # temp_path = F'weather_data/data-{date}.csv'
    # df.to_csv(temp_path,index=False)
    data_list = [
                        df.iloc[0,1],
                        df.iloc[0,2],
                        df.iloc[0,3],
                        df.iloc[1,1],
                        df.iloc[1,2],
                        df.iloc[1,3],
                        df.iloc[2,1],
                        df.iloc[2,2],
                        df.iloc[2,3],
                        df.iloc[3,1],
                        df.iloc[3,2],
                        df.iloc[3,3],
                        df.iloc[4,1],
                        df.iloc[4,2],
                        df.iloc[4,3],
                        df.iloc[4,4],
                        df.iloc[5,1],
                        df.iloc[5,2],
                        df.iloc[5,3],
                        df.iloc[5,4],
                        df.iloc[6,1],
                        df.iloc[6,2],
                        df.iloc[6,3],
                        df.iloc[7,1],
                        df.iloc[7,2],
                        df.iloc[7,3],
                        df.iloc[8,1],
                        df.iloc[8,2],
                        df.iloc[8,3],
                        date,
                        'Várzea Grande',
    ]
    building_df.loc[row_counter] = data_list
    # list_of_dfs.append(df)
    row_counter = row_counter + 1

# print(list_of_dfs)

building_df.to_csv('weather_data_Várzea_Grande_1999-2019.csv',index=False)
```
