```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

base_url = "https://www.wunderground.com/history/monthly/br/bel%C3%A9m/SBBE/date/"

# we are only trying this for one year
years = list(range(1999,2000))
months = list(range(1,3))

list_of_dates = []
for year in years:
  for month in months:
    date = F"{year}-{month}"
    print(date)
    list_of_dates.append(date)


# Set up options for the Chrome webdriver
chrome_options = Options()
# chrome_options.add_argument("--headless")  # Run in headless mode (no GUI)
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model

# Set up the Chrome webdriver
driver = webdriver.Chrome(options=chrome_options)

list_of_dfs = []
for date in list_of_dates:
    temp_url = base_url + date
    # Navigate to the web page you want to scrape
    
    imported_flag = False
    failed_atempts = 0
    while imported_flag == False and failed_atempts < 5:
      try:
        driver.get(temp_url)

        # Get the page source and pass it to Beautiful Soup for parsing


        sleep(3)

        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")


        table_div = soup.find_all(class_="summary-table")


        # Do your scraping here...
        print(table_div)

        table_for_df = table_div[0].find("table")

        # if this does not work it will throw an error
        df = pd.read_html(str(table_for_df))[0]

        imported_flag = True
        print(F"Successful import for date {date}moving into next season ")
      except ImportError:
        failed_atempts = failed_atempts + 1
        print(F"Failed to import date {date} attempt {failed_atempts}")
      

    print(df.head())

    # temp_path = F'weather_data/data-{date}.csv'
    # df.to_csv(temp_path,index=False)
    list_of_dfs.append(df)

print(list_of_dfs)
```

```python
import re

def extract_numerical_value(text):
    numerical_value = re.search(r'\d+', str(text))
    return float(numerical_value.group()) if numerical_value else None

# Assuming your DataFrame is named 'df'
columns_to_clean = ['Temperature', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']

for column in columns_to_clean:
    df[column + '_Cleaned'] = df[column].apply(extract_numerical_value)

# Print the updated DataFrame
df
```

```python
def averaging_day (df):

    condition_value = df['Condition'].value_counts().idxmax()
    wind_value = df['Wind'].value_counts().idxmax()
    # Assuming your DataFrame is named 'df'
    columns_to_clean = ['Temperature', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']

    avr_values = {'avr_Condition':condition_value, 
                  'avr_Wind': wind_value
                  }
    for column in columns_to_clean:
        df[column + '_Cleaned'] = df[column].apply(extract_numerical_value)

    for column in columns_to_clean:
        avr_values['avr_' + column] = df[column + '_Cleaned'].mean()
    
    return avr_values

# print(averaging_day(df))
# print(days_list)
days_list = [averaging_day(df)]

days_list.append(averaging_day(df))
print(days_list)
```
